```python
import threading
import time
import shutil

def backup_task():
    while True:
        if os.path.exists('state.cpt'):
            shutil.copy('state.cpt', 'state_backup.cpt')
            print(f"[{time.strftime('%H:%M:%S')}] Checkpoint backed up.")
        time.sleep(900) # 15 minutes

# Back groud running
threading.Thread(target=backup_task, daemon=True).start()
```
